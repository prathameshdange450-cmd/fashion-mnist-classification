```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
x_train_data=x_train/255
x_test_data=x_test/255
```

```python
classes=['T-shirt/top','Trouser',' Pullover',	'Dress', 'Coat', 'Sandal',	'Shirt', 'Sneaker',' Bag',	'Ankle boot']
```

```python
#one hot encoding is done beacuse the lables is in the words
from tensorflow.keras.utils import to_categorical
```

```python
y_train_categorical=to_categorical(
    y_train,num_classes=10
)
y_test_categorical=to_categorical(
    y_test,num_classes=10
)
```

```python
x_train_reshaped=x_train_data.reshape(len(x_train_data),28*28)
x_test_reshaped=x_test_data.reshape(len(x_test_data),28*28)
    
```

```python
model=keras.Sequential([
    keras.layers.Dense(500,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(optimizer='SGD',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train_reshaped,y_train_categorical,epochs=100)
```

```python
model.fit(x_train_reshaped,y_train_categorical)
```

```python
prediction=model.predict([x_test_reshaped,y_test_categorical])
```
